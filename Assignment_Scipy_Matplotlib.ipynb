# We have the min and max temperatures in a city In India for each months of the year.We would like to find a function to describe this and show it graphically, the dataset given below.
Task:
1.fitting it to the periodic function
2.plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25

Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
t_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
t_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
months = np.arange(1,13) #12 months of a year
```

```python
#print("avg max temp : ", np.average(t_max))
#print("avg max temp : ", np.average(t_min))
```

```python
plt.plot(months, t_max, 'ro')
plt.plot(months, t_min, 'bo')
plt.xlabel('Months')
plt.ylabel('Min and max temperature')
```

```python
def yearly_mytemps1(x, p0,p1,p2,p3):
    return (p0+p1*x**1+p2*x**2+p3*x**3)

res_max1, cov_max1 = curve_fit(yearly_mytemps1, months, t_max) #fitting max temps
res_min1, cov_min1 = curve_fit(yearly_mytemps1, months, t_min) #fitting the min temps

days = np.linspace(0, 12, num=365)

a=plt.figure()
axes= a.add_axes([0.1,0.1,0.8,0.8])
plt.plot(months, t_max, 'ro') #plotting the max temps in red color
plt.plot(days, yearly_mytemps1(days, *res_max1), 'r-') #fitting the max temp 
plt.plot(months, t_min, 'bo') #plotting the min temps in blue color
plt.plot(days, yearly_mytemps1(days, *res_min1), 'b-') #fitting the min temp
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
axes.set_ylim([10,60])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic=pd.read_csv(url)
titanic.head(5)
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic.shape
```

```python
titanic.dropna(subset=['sex','age','fare'], inplace = True) #removed all the NAN values in the three columns age, fare and sex
print(titanic.shape[0])
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

values =titanic.sex.value_counts()
colors = ['b', 'r']
labels = ['male', 'female']
explode = (0.12, 0)
plt.pie(values, autopct='%1.1f%%', colors=colors, labels= values, explode=explode,counterclock=False, shadow=True,pctdistance=0.751, labeldistance=1.130 )
plt.title('Male/Female Proportion with counts and percentage where out of 1310 only 1309 are plotted as one is a row with NAN')
plt.legend(labels, loc=4)
plt.show()
```

```python
titanic['sex'] = titanic['sex'].map(dict(zip(['male','female'],[0,1]))) #mapped the male and female as 0 and 1 for ease of plotting 
titanic['sex'].head(10)
```

```python
#scatter plot by using matplotlib 

figure(figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')
scatter = plt.scatter(titanic.age, titanic.fare, c=titanic.sex, cmap = plt.cm.RdYlGn )
plt.xlabel("age")
plt.ylabel("fare")
plt.title('A scatterplot with the Fare paid and the Age, differ the plot color by gender')
plt.legend(*scatter.legend_elements(),loc="upper right", title = "sex - Male 0, Female - 1")

plt.show()
```
